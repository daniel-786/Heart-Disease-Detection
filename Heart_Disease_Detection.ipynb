#Heart Disease Detection

**Import Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data Collection and Processing**

```python
# loading dataset
heart_dataset = pd.read_csv("/content/heart_disease_data.csv")
```

```python
# showing dataset
heart_dataset.head()
```

```python
# plot testing with single column

num_rows, num_cols = 1, 5
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 5))

col = heart_dataset["age"]
sns.countplot(x = col, ax = axs[0])
axs[0].set_title("Count Plot for age")

plt.tight_layout()
plt.show()
```

```python
# automating for full dataset

num_rows, num_cols = 5, 3
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 20))

for index, col_name in enumerate(heart_dataset.columns):
    #if(col_name != "trestbps" and col_name != "chol" and col_name != "thalach" and col_name != "oldpeak"):
        col = heart_dataset[col_name]

        row_index = index // num_cols
        col_index = index % num_cols

        sns.countplot(x = col, ax = axs[row_index, col_index])
        axs[row_index, col_index].set_title(f"Plot {col_name}")


plt.tight_layout()
plt.show()
```

```python
# separating features and target
X = heart_dataset.drop(["target"], axis = 1)
Y = heart_dataset["target"]
```

```python
print(X)
print(Y)
```

**Split Train and Test data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print("Total data: ", X.shape)
print("Train data: ", X_train.shape)
print("Test data: ", X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

**Model Training using Logistic Regression**

```python
model = LogisticRegression(max_iter = 1000)
```

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Accuracy on training data: ", training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy on training data: ", test_data_accuracy)
```

**Predictive System**

```python
# taking a sample value for test
X_test.iloc[18].values
```

```python
# checking sample label
Y_test.iloc[18]
```

```python
input_data = X_test.iloc[18].values

# reshape the numpy array as we're predicting for one instance
input_data_reshaped = input_data.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] == 0):
    print("Diagnosis shows no Heart Disease")
else:
    print("Diagnosis detects Heart Disease")
```
